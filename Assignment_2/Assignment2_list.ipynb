```python
# Function to append a number to the main list
def list_create(number, main_list):
    main_list.append(number)

# Function to calculate the sum of all elements in the list
def my_sum(main_list):
    total_sum = 0
    for i in range(len(main_list)):
        total_sum = main_list[i] + total_sum
    return total_sum

# Function to calculate the average of the elements in the list
# Error could occur if main_list is empty, so we handle division by zero
def my_avg(totalsum, main_list):
    if len(main_list) == 0:
        # Prevent division by zero
        print("Error: Division by zero - List is empty.")
        return 0
    avg = totalsum / len(main_list)
    return avg

# Function to find the maximum value in the list
# Error can occur if the list is empty, but we assume the list will have at least one element
def maximum(main_list):
    if len(main_list) == 0:
        # Handle case where list is empty to avoid index error
        print("Error: Cannot find the maximum of an empty list.")
        return None
    max_value = main_list[0]  # Start with the first element
    for i in range(len(main_list)):
        if max_value < main_list[i]:
            max_value = main_list[i]
    return max_value

# Function to find the minimum value in the list
# Error can occur if the list is empty, similar to the maximum function
def minimum(main_list):
    if len(main_list) == 0:
        # Handle case where list is empty to avoid index error
        print("Error: Cannot find the minimum of an empty list.")
        return None
    min_value = main_list[0]  # Start with the first element
    for i in range(len(main_list)):
        if min_value > main_list[i]:
            min_value = main_list[i]
    return min_value

# Function to calculate the cumulative sum of the elements in the list
def cumulative_sum(main_list):
    total_sum = 0
    cumulative_list = []
    for i in range(len(main_list)):
        total_sum = main_list[i] + total_sum
        cumulative_list.append(total_sum)
    return cumulative_list

# Initialize an empty list called main_list
main_list = []

# Input loop to collect numbers from the user
while True:
    number = input("Enter the integer number you want to store in the list or enter 'stop' to finish: ")
    
    # Convert the input to uppercase to avoid case-sensitive errors
    if number.upper() == "STOP":  # Check if the user wants to stop input
        print("No more inputs")
        break
    try:
        number = int(number)  # Convert the input to an integer
    except ValueError:
        # Handle case where the input is not an integer
        print("Error: Please enter a valid integer.")
        continue
    list_create(number, main_list)

# Print the list of numbers
print("List of numbers:", main_list)

# Check if the list is empty before performing operations
if len(main_list) == 0:
    print("Error: The list is empty, no operations can be performed.")
else:
    # Calculate and print the total sum of the list
    totalsum = my_sum(main_list)
    print("Total Sum:", totalsum)

    # Calculate and print the average of the list
    avg = my_avg(totalsum, main_list)
    print("Average:", avg)

    # Find and print the maximum value in the list
    max_value = maximum(main_list)
    print("Maximum Value:", max_value)

    # Find and print the minimum value in the list
    min_value = minimum(main_list)
    print("Minimum Value:", min_value)

    # Calculate and print the cumulative sum of the list
    cumulative_list = cumulative_sum(main_list)
    print("Cumulative Sum List:", cumulative_list)
```
